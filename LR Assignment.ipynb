```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
from sklearn.impute import SimpleImputer
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
# get the dataset
test = pd.read_csv("test.csv")
test
```

```python
#split the dataset 
x=test.iloc[:,0].values.reshape(-1,1)
y=test.iloc[:,1].values.reshape(-1,1)
```

```python
# split the data set into x train,x test, y train, y test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# from linear regression improt linear regression functions
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# predict the value form test set
predictvalue = lr.predict(x_test)
predictvalue
```

```python
compare = pd.DataFrame(x_test,columns=['x'])
```

```python
# compare the avaliable value and predict value
compare["AVLDATA"]=y_test
compare['Predictdata']=predictvalue
compare
```
