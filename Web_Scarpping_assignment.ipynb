```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
req = rq.get(url)
req
```

```python
html = BeautifulSoup(req.text)
html
```

```python
title = html.select("._4rR01T")
title
```

```python
cost = html.select("._1_WHN1")
cost
```

```python
pros = html.select(".rgWa7D:nth-child(1)")
pros
```

```python
disp = html.select(".rgWa7D:nth-child(5)")
disp
```

```python
details = pd.DataFrame(title,columns=["Name Of The Product"])
details["Processer"]=pros
details["Display"]=disp
details["Cost"]=cost
details
```

```python
details.to_csv("Laptop_Details.csv")
```
